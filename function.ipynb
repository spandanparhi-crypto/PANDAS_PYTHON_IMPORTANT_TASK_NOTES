```python
# make the function reverse the value 

def reverse_num(num):
    
    
    reverse = 0
    
    while num > 0 :
        
        digit  = num % 10
        reverse =  reverse * 10 + digit
        num  = num // 10
        

    return reverse

```

```python
reverse_num(1234)
```

```python
def check_min(*number):
        mininum  = number[0]
    
    for num in number:
        if num < mininum:
            mininum  = num
            
    return mininum

```

```python
check_min(10,20,1)
```

```python
def small_num(*numbers):
    small = numbers[0]

    for num in numbers:
        if num < small:
            small = num

    return small

```

```python
print(small_num(100,25,325,5899))
```

```python
# count the even number 
 
def count_even_num(*numbers):
    
    count = 0
    
    for num in numbers:
        if num % 2 == 0:
            count += 1
            
    return  count       
```

```python
count_even_num(1,2,3,4,5,6)
```

```python
# shallow copy
#import lib

import copy

list1 = [[1,3],[4,6]]  # original list
       # () 
list2 = copy.copy(list1)

# update the value 
list2[1][0]= 8

print("Orginal List:" , list1)
print("Copy List:" , list2)
```

```python
# create one lists  : three inner list inside the outer list : (2,1)
list1 = [[1,2],[3,4],[5,6]] # original list

# create the copy list

list2 = copy.copy(list1)

# update the value (2,1)
list2[2][1] = 45

print("Orginal List:" , list1)
print("Copy List:" , list2)
```

```python
import copy

original = [1, 2, [3, 4]]

shallow = copy.copy(original)

shallow[0] = 100

print("Original:", original)
print("Shallow :", shallow)
```

```python
#  check outer list address
import copy

original = [1, 2, [3, 4]]

shallow = copy.copy(original)

print(id(original))
print(id(shallow))
```

```python
# Check Nested List Address
import copy

original = [1, 2, [3, 4]]

shallow = copy.copy(original)

print(id(original[2]))
print(id(shallow[2]))
```

```python
# original list
list1 = [1,2,[4,5]]

# shallow copy list 

list2 = copy.copy(list1)

# update the list2
list2[0] = 49

print("Orginal Lists",list1)
print("Copy List Lists",list2)
```

```python
# deep copy :

list1 = [[1,2],[3,4]]  # original List

# create the new list

list2 = copy.deepcopy(list1)

#update the list

list2[1][1] = 56

print("Orginal Lists",list1)
print("Copy List Lists",list2)
```

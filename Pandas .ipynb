```python
# handled the outlier 

df = pd.DataFrame({"Age" : [32,43,23,45,54,23,45,129,31,180,10,2]})
df
```

```python
# detect the outlier 

Q3 =df['Age'].quantile(0.75)
Q1 = df['Age'].quantile(0.25)

# find the IQR

IQR = Q3 - Q1


# find the lowe bound and upper bound 

lower_bound = Q1 - 1.5*IQR

upper_bound = Q3 + 1.5*IQR


# display the outlier 

outlier=  df[(df["Age"] < lower_bound) | (df["Age"] > upper_bound)]
```

```python
# Handled the outlier using clip method 
df["Age"]=df['Age'].clip(lower_bound,upper_bound)
```

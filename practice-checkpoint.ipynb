# 1. Print Numbers from 1 to 20

```python
num = 1

while num <= 20:
    print(num)
    num += 1
```

# 2. Print Even Numbers

```python
num = 1

while num <= 50:
    if num % 2 == 0:
        print(num)

    num += 1
```

# 3. Sum of Natural Numbers

```python
n = int(input("Enter number: "))

i = 1
total = 0

while i <= n:
    total += i
    i += 1

print("Sum =", total)
```

# 4. Reverse a Number

```python
num = int(input("Enter number: "))

reverse = 0

while num > 0:
    digit = num % 10
    reverse = reverse * 10 + digit
    num = num // 10

print("Reverse =", reverse)
```

# 5. Count Digits in a Number

```python
num = int(input("Enter number: "))

count = 0

while num > 0:
    num = num // 10
    count += 1

print("Total digits =", count)
```

# 6. Palindrome Number

```python
num = int(input("Enter number: "))

original = num
reverse = 0

while num > 0:
    digit = num % 10
    reverse = reverse * 10 + digit
    num = num // 10

if original == reverse:
    print("Palindrome")
else:
    print("Not Palindrome")
```

# 7. Armstrong Number

```python
num = int(input("Enter number: "))

original = num
total = 0

while num > 0:
    digit = num % 10
    total += digit ** 3
    num = num // 10

if total == original:
    print("Armstrong Number")
else:
    print("Not Armstrong")
```

# 11. Fibonacci Series

```python
n = int(input("Enter terms: "))

a = 0
b = 1
count = 1

while count <= n:
    print(a)

    c = a + b
    a = b
    b = c

    count += 1
```

# 12. Factorial of Number

```python
num = int(input("Enter number: "))

fact = 1

while num > 0:
    fact *= num
    num -= 1

print("Factorial =", fact)
```

# 13. Menu Driven Calculator

```python
while True:

    print("\n1.Add")
    print("2.Subtract")
    print("3.Multiply")
    print("4.Divide")
    print("5.Exit")

    choice = int(input("Enter choice: "))

    if choice == 5:
        print("Exit")
        break

    num1 = int(input("Enter first number: "))
    num2 = int(input("Enter second number: "))

    if choice == 1:
        print("Answer =", num1 + num2)

    elif choice == 2:
        print("Answer =", num1 - num2)

    elif choice == 3:
        print("Answer =", num1 * num2)

    elif choice == 4:
        print("Answer =", num1 / num2)

    else:
        print("Invalid Choice")
```

# 14. ATM System

```python
balance = 10000

while True:

    print("\n1.Check Balance")
    print("2.Deposit")
    print("3.Withdraw")
    print("4.Exit")

    choice = int(input("Enter choice: "))

    if choice == 1:
        print("Balance =", balance)

    elif choice == 2:
        amount = int(input("Enter deposit amount: "))
        balance += amount
        print("Updated Balance =", balance)

    elif choice == 3:
        amount = int(input("Enter withdraw amount: "))

        if amount <= balance:
            balance -= amount
            print("Remaining Balance =", balance)
        else:
            print("Insufficient Balance")

    elif choice == 4:
        print("Thank You")
        break

    else:
        print("Invalid Option")
```

#  16. Leap Year Checker

```python
while True:

    year = input("Enter year or stop: ")

    if year == "stop":
        break

    year = int(year)

    if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
        print("Leap Year")

    else:
        print("Not Leap Year")
```

# 18. Prime Numbers Between 1 to 100

```python
num = 2

while num <= 100:

    i = 2
    prime = True

    while i < num:

        if num % i == 0:
            prime = False
            break

        i += 1

    if prime:
        print(num)

    num += 1
```

# 19. Student Marks Average

```python
total = 0
count = 0

while True:

    marks = int(input("Enter marks: "))

    if marks < 0:
        break

    total += marks
    count += 1

average = total / count

print("Average =", average)
```

# 20. Login System with 3 Attempts

```python
username = "admin"
password = "1234"

attempts = 3

while attempts > 0:

    user = input("Enter username: ")
    pwd = input("Enter password: ")

    if user == username and pwd == password:
        print("Login Successful")
        break

    else:
        attempts -= 1
        print("Wrong Credentials")
        print("Remaining Attempts =", attempts)

if attempts == 0:
    print("Account Blocked")
```

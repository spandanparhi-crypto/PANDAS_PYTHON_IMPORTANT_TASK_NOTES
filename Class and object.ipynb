```python
# with abstraction


from abc import ABC,abstractmethod

class Animal(ABC):
    
    
    # method 1:
    @abstractmethod   # decorator
    def speak(self):
        pass
    
    # method 2:
    @abstractmethod
    def move(self):
        pass
    
    
    # method 3:
    def breating(self):
        print("Breathing Animal")
        
    # method 
    def show(self):
        self.speak()
        self.move()
        
        
        
        
# class child class
class Cat(Animal):
    
    def speak(self):
        print("Mewoooooo")
        
    def move(self):
        print("🐾")
        

# child class

class Dog(Animal):
        
    def speak(self):
        print("Bhooouuuuuu")
        
    def move(self):
        print("🐾")
        
    

# create the object

cat = Cat()
dog = Dog()


# show using cat
dog.show()

    
    
```

```python

# parent class
class Zomato(ABC):
    # constructor
    def __init__(self,f_name,f_price):
        self.__food_name = f_name
        self.__food_price = f_price
    
    
    # accessor : food name 
    def get_food_name(self):
        return self.__food_name
    
    # accessor : food price
    def get_food_price(self):
        return self.__food_price
    
    
    # abstract method 
    @abstractmethod 
    def food_prepare(self):
        pass
    
    # abstract method
    def food_delivered(self):
        pass
    
    # instace method  show()
    def show(self):
        print(self.__food_name)
        print(self.__food_price)
        self.food_prepare()
        self.food_delivered()
        
        
# child 1:
class Pizza(Zomato):
    def food_prepare(self):
        print("Pizza is prepared .......")
    
    
    def food_delivered(self):
        print("Pizza is delivered .......")
        
# child 2:
class Birayni(Zomato):
    def food_prepare(self):
        print("Birayni is prepared .......")
      
    def food_delivered(self):
        print("Birayni is delivered .......")
        
        
# creath child class object 
pizza  = Pizza("Pizza",350) 

#pizza.show()

# create the child class 

biryani = Birayni("Veg Birayani",250)
biryani.show()        
    
    

            
```

```python
class University:

    def __init__(self, University_Name, City, Number_Of_Colleges, **kwargs):
        super().__init__(**kwargs)

        self.University_Name = University_Name
        self.City = City
        self.Number_Of_Colleges = Number_Of_Colleges


class College(University):

    def __init__(self, College_Name, **kwargs):
        super().__init__(**kwargs)

        self.College_Name = College_Name


class School(University):

    def __init__(self, School_Name, **kwargs):
        super().__init__(**kwargs)

        self.School_Name = School_Name


class Student(College, School):

    def __init__(
        self,
        University_Name,
        City,
        Number_Of_Colleges,
        College_Name,
        School_Name,
        Student_Name
    ):

        super().__init__(
            University_Name=University_Name,
            City=City,
            Number_Of_Colleges=Number_Of_Colleges,
            College_Name=College_Name,
            School_Name=School_Name
        )

        self.Student_Name = Student_Name

    def Display_Details(self):

        print("University Name :", self.University_Name)
        print("City :", self.City)
        print("Number Of Colleges :", self.Number_Of_Colleges)
        print("College Name :", self.College_Name)
        print("School Name :", self.School_Name)
        print("Student Name :", self.Student_Name)


# Create Object

Std1 = Student(
    "Mumbai University",
    "Mumbai",
    850,
    "KJ Somaiya College",
    "Somaiya School",
    "Sumit"
)

Std1.Display_Details()
```

```python
class Credit_card:
    def pay(self,amount):
        print(f"Paying {amount} via Credit Card")
        
class Debit_card:
    def pay(self,amount):
        print(f"Paying {amount} via Debit Card")
        
        
class UPI_mode:
    def pay(self,amount):
        print(f"Paying {amount} via UPI Mode")
        
# create the duck typing

def make_payments(payments_mode,amount):
    payments_mode.pay(amount)
    
c =Credit_card()
d =Debit_card()
u =UPI_mode()


# calling the duck typinh 

make_payments(c,1000)
make_payments(d,2000)
make_payments(u,3000)
    
print("─" * 30)     


# another techniques to call the duck type 

payments = [Credit_card(),Debit_card(),UPI_mode()]

for payment in payments:
    payment.pay(100)
    
```

```python
# type 1: duck tping

# class 1
class Credit_card:
    
    # method 1:
    def pay(self,amount):
        print(f"Paying {amount} via Credit Card")
        
        
# class 2
class Debit_card:
    
    #method 2 :
    def pay(self,amount):
        print(f"Paying {amount} via Debit Card")
        
        
# class 3:
class Phone_pay:
    
    # method 3:
    def pay(self,amount):
        print(f"Paying {amount} via Phone Pay")
        
        
        
        
        
# duck typing

# make the function 

def make_payments(payement_mode,amount):
    return payement_mode.pay(amount)
  

# create the class object 

c =Credit_card()

d = Debit_card()

p = Phone_pay()


# call the function

make_payments(c,100)
make_payments(d,200)
make_payments(p,300)

   
print("-"*20)

# make the class list 
payments =[Credit_card(),Debit_card(),Phone_pay()]

for payement in payments:
    payement.pay(100)
```

```python
# example 2 : 


class Circle:
    
    # constructor 
    def __init__(self,r):
        self.Radius = r
        
        
    # method 1
    def draw(self):
        print(f"Draw the circle using {self.Radius}")
        
        
    # method 2:
    def area(self):
        print(f"Area of Cricle is :  ",3.14*(self.Radius**2))
        
        
# class 2:

class Rectangle:
    
    # constructor
    def __init__(self,l,w):
        self.Length = l
        self.Width  = w
        
    # method 1
    def draw(self):
        print(f"Draw the Rectangle using {self.Length}")
        
        
    # method 2:
    def area(self):
        print(f"Area of Rectangle is : {self.Length * self.Width} ")
        
        
# class 3:

class Triangle:
    # constructor
    def __init__(self,b,h):
        self.Base = b
        self.Height  = h
        
    # method 1
    def draw(self):
        print(f"Draw the Triangle using {self.Base}")
        
        
    # method 2:
    def area(self):
        print(f"Area of Triangle is : {0.5 *self.Base * self.Height} ")
        
        
        
# make_shape functionn

#shapes = [Circle(),Rectangle(),Triangle()]


def show_shape(shape):
    
    shape.draw() ,shape.area()
    print("─" * 30)
    
    
# create the class object 


c = Circle(4)
r = Rectangle(2,5)
t = Triangle(3,5)


# call the function 

show_shape(c)
show_shape(r)
show_shape(t)

    
    
        
        
    
    
    

    
    
    
    
```

```python
# type : operator overloading

class Student:
    
    # const
    def __init__(self,a,b):
        self.A = a
        self.B = b
        

    # dunder method
    def __add__(self,other):
        return self.A +other.B
    
    def __sub__(self,other):
        return self.A - other.B
    
    # method : mul
    def __mul__(self,other):
        return self.A * other.B        
        
ob1 = Student(1,2)
ob2 = Student(3,4)
ob3 = Student(5,6)


print(ob1 + ob3 )
print(ob1 - ob1 )
print(ob1 * ob3 )
```
